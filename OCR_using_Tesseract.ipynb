```python
from PIL import Image #importing module from 'pillow', for image processing
import pytesseract #importing python API of Tesseract (Google's library), for extracting text from images
```

```python
#loading document/image
img = Image.open('doc.jpg')
img
```

```python
#converting image to grayscale(3 channels to 1 channel) for better accuracy
gray_image = img.convert('L') #'L'=Luminance/grayscale, '1'=Binarize(pixel value 0 or 255)
gray_image
```

```python
#extracting text from image using python API of Tess 
text = pytesseract.image_to_string(gray_image)
print(text)
```
